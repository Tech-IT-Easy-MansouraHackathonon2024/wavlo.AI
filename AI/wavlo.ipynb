```python
from pydub import AudioSegment
import speech_recognition as sr
import time

# Function to split audio into chunks
def split_audio(audio_file, chunk_duration):
    sound = AudioSegment.from_wav(audio_file)
    chunks = []
    for i in range(0, len(sound), chunk_duration * 1000):  # Convert duration to milliseconds
        chunks.append(sound[i:i + chunk_duration * 1000])
    return chunks

# Function to recognize text from an audio chunk
def recognize_audio_chunk(chunk, recognizer):
    try:
        with sr.AudioFile(chunk.export("temp.wav", format="wav")) as source:
            audio_data = recognizer.record(source)
            return recognizer.recognize_google(audio_data, language="ar-EG")  # Arabic recognition
    except sr.UnknownValueError:
        return "Unable to recognize the audio."
    except sr.RequestError as e:
        return f"Error with the recognition service: {e}"

# Main process
audio_file = "/kaggle/input/fatihahahh/abdul-rahman-al-sudais-001-al-fatiha-3218-8020.wav"
chunk_duration = 30  # Split audio into chunks of 30 seconds

recognizer = sr.Recognizer()
chunks = split_audio(audio_file, chunk_duration)

print("Number of chunks:", len(chunks))

for i, chunk in enumerate(chunks):
    print(f"Processing chunk {i + 1}...")
    text = recognize_audio_chunk(chunk, recognizer)
    
    # Display words one by one with delay
    for word in text.split():
        print(word, end=" ", flush=True)
        time.sleep(0.3)  # Adjust the delay time in seconds
    print("\n")  # Move to the next line for the next chunk
```
